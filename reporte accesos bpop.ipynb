```python
import pandas as pd
from natsort import natsorted
import glob
from datetime import date
```

```python
file_path = "D:\\Documents\\Cristian\\Proyectos varios\\Proyectos Python\\Web Scrapping\\Usuarios PBI\\"

file_paths = natsorted(glob.glob(file_path + "*accesos BPOP*.xlsx", recursive=True))
```

```python
df = pd.DataFrame()

for f in file_paths:
    #print(f)
    accesos = pd.read_excel(f, 'Sheet1')
    #print(accesos.columns)
    df = pd.concat([df, accesos], ignore_index = True)
```

```python
df.head()
```

```python
df_t = df[(df["Permisos"] == "Read") & (~df["Correo"].str.contains("@AvalDigitalLabs.com", regex = False))].drop(["Unnamed: 0", "iniciales", "Area", "Entidad"], axis = 1).drop_duplicates()
df_t["Url"] = df_t["Url"].str.replace("/permission/report/2", "/reports").str.replace("/directAccess?experience=power-bi", "/ReportSection?experience=power-bi")
```
